**1. 자료구조 Dictionary를 이용하여, if 문 없이 연도를 입력했을 때 윤년/평년을 판별하여 출력하는 프로그램을 생성하시오.**

```python
year = int(input("연도 입력: "))

leap_year_dict = {True: "윤년", False: "평년"}
result = leap_year_dict[(year % 4 == 0 
                    and year % 100 != 0) 
                    or (year % 400 == 0)]
print(result)
```

**2. 키와 몸무게를 입력받아 BMI 수치를 계산하는 프로그램을 제작하고, 새로운 값을 입력할 때마다, result_list에 결과를 추가하시오.**

* 조건 : 새로운 값을 입력 할 때 마다 기존의 데이터가 삭제되지 않고 추가되어야 합니다.



```python
result_list = []

while True:
    height = float(input("키 입력<cm>: "))
    weight = float(input("몸무게 입력: "))
    
    bmi = weight / (height * 0.01) ** 2
    
    print(f"BMI 수치는: {bmi:.2f}")

    result_list.append(bmi)

    option = int(input("그만하려면 -1 입력: "))
    if option == -1:
        break

format_result_list = [f"{b:.2f}" for b in result_list]
print(f"BMI 수치 리스트: {format_result_list}")
```

**3. A 학생의 과목별 점수가 아래와 같을 때, 이 점수를 딕셔너리 형태로 선언하고, 평균을 계산하오**

* A : 국어 75, 영어 53, 수학 90, 사회 46, 한국사 60, 세계사 70, 과학, 34
* 과목명은 Key값, 점수는 Value 값
     

```python
score_dict = {'국어': 75, '영어': 53, '수학': 90,
         '사회': 46, '한국사': 60, '세계사': 70,
         '과학': 34}
avg = sum(score_dict.values()) / len(score_dict)
print(f"과목 평균: {avg:.2f}")
```

**4. 어떤 수를 입력했을 때, 입력한 수가 홀수이면 True, 짝수이면 False를 출력하는 프로그램을 제작하시오**

* if문을 사용 하지 않고 제작 

```python
def odd_even(num):
    return num % 2

odd_even_dict = {0: "짝수", 1: "홀수"}

num = int(input("숫자 입력: "))

result = odd_even_dict[odd_even(num)]
print(result)
```

**5. For 반복문을 이용하여, 1부터 입력받은 수 까지 제곱 합을 구하는 프로그램을 만드시오.**

Ex ) 1² + 2² + 3² + 4² + ...  + n²

```python
num = int(input("숫자 입력: "))

sum = 0
for i in range(1, num + 1):
    sum += i ** 2

print(sum)
```

**11. 유입경로(Channel)에서 '영업방판'으로 유입된 고객들을 분석하고자 한다.'영업방판'으로 유입된 고객들의 월랜탈비용(Amount_Month)의 평균(mean)과 중앙값(median)의 차이를 계산 하시오.**

```python
import pandas as pd
df1 = pd.read_csv('01_Data.csv')

door_to_door = df1[df1['Channel'] == '영업방판']

difference = door_to_door['Amount_Month'].mean() -\
             door_to_door['Amount_Month'].median()

#3422.79
print("{:.2f}".format(difference))
```

**12. 유입경로(Channel)에서 '영업방판'으로 유입된 고객들 중, 제품군(Product_Type) 별 월랜탈비용(Amount_Month)의 합(sum)을 계산하여 result3.xlsx 파일로 저장하시오.**

```python
import pandas as pd
df1 = pd.read_csv('01_Data.csv')

door_to_door = df1[df1['Channel'] == '영업방판']

result = door_to_door.groupby('Product_Type')['Amount_Month'].sum()\
                                                             .reset_index()
result.to_excel('result3.xlsx', index = False)

print("result3.xlsx 파일로 저장완료")
```

**15. 제품군(Product_Type)중 빈도수가 가장 높은 상위 3개의 항목을 추출하여, df2로 선언한 뒤, 계약일자(Datetime)에 따른 월렌탈비용(Amount_Month)의 합을 계산해 result5.xlsx파일로 저장하시오.**

```python
import pandas as pd
df1 = pd.read_csv('01_Data.csv')

top3 = df1['Product_Type'].value_counts().head(3).index
df2 = df1[df1['Product_Type'].isin(top3)]

result = df2.groupby('Datetime')['Amount_Month'].sum()\
                                                .reset_index()
result.to_excel('result5.xlsx', index = False)

print("result5.xlsx 파일로 저장완료")
```
