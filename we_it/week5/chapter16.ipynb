```python
import pandas as pd
```

```python
df_raw = pd.read_csv('exam.csv')
df = df_raw.head(10)
df
```

```python
#nclass가 1이면 추출
df[df['nclass'] == 1]
```

```python
#수학점수가 80점 이상이면 추출
df[df['math'] >= 80]
```

```python
#nclass가 1이면서 수학점수가 50점 이상
df[(df['nclass'] == 1) & (df['math'] >= 50)]
```

```python
#수학 점수가 50점 미만이거나 영어 점수가 50점 미만
df[(df['math'] < 50) | (df['english'] < 50)]
```

```python
df = pd.DataFrame({
    'var1' : [1,2,3],
    'var2' : [4,5,6]
})
```

```python
df
```

```python
#인덱스 문자열 지정하기
df = pd.DataFrame({
    'var1' : [1,2,3],
    'var2' : [4,5,6]},
    index = ['kim','lee','park']
)
```

```python
df
```

```python
#인덱스 문자열을 지정해서 행 추출하기
```

```python
#kim 행 추출
df.loc['kim']
```

```python
#여러 행 추출하기
df.loc[['kim','park']]
```

```python
#인덱스 문자열을 지정해 열 추출하기
```

```python
#lee 행의 var1열 추출
df.loc['lee','var1']
```

```python
#kim, lee 행의 var2열 추출
df.loc[['kim','lee'], 'var2']
```

```python
#df.loc[ ]이용해서 열 추출 시, 반드시 행 추출 기준도 함께 입력해야 함
df.loc[ ,'var1']
```

```python
#모든 행을 추출하려면 : 을 입력해야 함
df.loc[:, 'var1']
```
