```python
import pandas as pd
```

```python
import numpy as np
```

```python
import seaborn as sns
```

```python
#문재인 대통령 대선 출마 선언문 파일 열고 불러오기
moon = open('speech_moon.txt',encoding = 'UTF-8').read()
```

```python
moon
```

```python
import re
```

```python
#불필요한 문자(특수문자/한자/공백) 제거하기
moon = re.sub('[^가-힣]', ' ', moon)
```

```python
moon
```

```python
##명사 추출하기
```

```python
import konlpy
```

```python
#hannanum 만들기
```

```python
hannanum = konlpy.tag.Hannanum()
```

```python
#명사 추출하기
hannanum.nouns("대한민국의 영토는 한반도와 그 부속도서로 한다")
#"대한민국의 영토는 한반도와 그 부속도서로 한다"라는 문장에서 명사만을 골라서 추출해줌
```

```python
#연설문에서 명사 추출하기
nouns = hannanum.nouns(moon)
```

```python
nouns
```

```python
#cf) hannanum.nouns()의 출력결과는 리스트임
```

```python
#데이터프레임으로 변환
df_word = pd.DataFrame({'word' : nouns})
```

```python
df_word
```

```python
#단어빈도표 만들어서 연설문에서 어떤 단어를 많이 사용했는지 알아보기
#한글자로 된 단어는 의미없는 경우 많으므로 제거
```

```python
#글자 수 추가
df_word['count'] = df_word['word'].str.len()
```

```python
df_word
```

```python
#두 글자 이상 단어만 남기기
df_word = df_word.query('count>=2')
```

```python
df_word.sort_values('count')
```

```python
#단어의 사용 빈도 구하기
df_word = df_word.groupby('word',as_index = False)\
                .agg(n=('word','count')) \
                .sort_values('n', ascending=False)
#(순서대로) 단어별 분리
#빈도 구하기
#내림차순 정렬
```

```python
df_word
```

```python
#출력 결과,
#'나라','일잘','국민' 등 단어를 많이 사용함
```

```python
#단어 빈도 상위 20개 추출
top20 = df_word.head(20)
```

```python
top20
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({
    'font.family' : 'Malgun Gothic',
    'figure.dpi' : '120',
    'figure.figsize' : [6.5,6]
})
#(순서대로) 한글폰트
#해상도
#가로 세로 크기
```

```python
#막대 그래프 만들기
sns.barplot(data=top20, y = 'word', x = 'n')
```

```python
#출력 결과, 
#'일자리','복지국가' 등의 단어를 보면
#문재인 대통령이 연설에서 일자리 문제를 해결하고 복지국가를 지향하겠다는 의사를 표현했음을 알 수 있음
```

```python
#워드 클라우드 : 단어의 빈도를 구름 모양으로 표현한 그래프
#             : 단어는 키(key), 빈도는 값(value)으로 구성된 딕셔너리 자료구조 이용해서 만듦
```

```python
font = 'DoHyeon-Regular.ttf' #한글 폰트 설정하기
#DoHyeon-Regular.ttf는 배달의민족 도현체 폰트파일
```

```python
#데이터프레임을 딕셔너리로 변환
dic_word = df_word.set_index('word').to_dict()['n']
```

```python
dic_word
```

```python
from wordcloud import WordCloud
```

```python
#wc 만들기
wc = WordCloud(
    random_state=1234,
    font_path = font,
    width = 400,
    height = 400,
    background_color = 'white'
)
#(순서대로) 난수 고정 // 난수 : 무작위로 생성한 수 
#폰트 설정
#가로크기
#세로 크기
#배경색
```

```python
#워드 클라우드 만들기
img_wordcloud = wc.generate_from_frequencies(dic_word)

#워드 클라우드 출력하기
plt.figure(figsize = (10,10))
plt.axis('off')
plt.imshow(img_wordcloud)
```

```python
#mask 만들기
#cf) mask는 이미지 파일을 이용해서 원하는 모양의 워드클라우드 만들 수 있게 함
```

```python
import PIL
icon = PIL.Image.open('cloud.png')

img = PIL.Image.new('RGB', icon.size, (255,255,255))
img.paste(icon,icon)
img = np.array(img)
```

```python
#워드 클라우드 만들기
wc = WordCloud(
    random_state = 1234,
    font_path = font,
    width = 400,
    height = 400,
    background_color = 'white',
    mask = img
)

#(순서대로) 난수 고정
#폰트 설정
#가로 크기
#세로 크기
#배경색
#mask 설정

img_wordcloud = wc.generate_from_frequencies(dic_word)

plt.figure(figsize = (10,10))
plt.axis('off')
plt.imshow(img_wordcloud)

#(순서대로) 가로, 세로 크기 설정
#테두리 선 없애기
#워드 클라우드 출력
```

```python
wc = WordCloud(
    random_state=1234, #난수 고정
    font_path = font, #폰트설정
    width = 400, #가로크기
    height = 400, #세로크기
    background_color = 'white', #배경색
    mask = img, #mask설정
    colormap = 'inferno' #컬러맵 설정
)
```

```python
#워드 클라우드 만들기
img_wordcloud = wc.generate_from_frequencies(dic_word)
```

```python
#워드 클라우드 출력하기
plt.figure(figsize=(10,10)) #가로, 세로 크기 설정
plt.axis('off') #테두리 선 없애기
plt.imshow(img_wordcloud) #워드 클라우드 출력
```

```python
##### 10-2) 기사 댓글 텍스트 마이닝
```

```python
#네이버 뉴스 기사 댓글을 이용한 텍스트 마이닝
```

```python
df = pd.read_csv('news_comment_BTS.csv', encoding = 'UTF-8')
```

```python
#데이터 살펴보기
df.info()
```

```python
#reply변수 : 댓글
#reply변수에서 불필요한 문자 제거 후, 한글만 남긴 다음 일부 출력해서 내용확인하자
```

```python
#불필요한 문자 제거 // 한글만 남기기
df['reply'] = df['reply'].str.replace('[^가-힣]', ' ', regex = True)
```

```python
df['reply'].head()
```

```python
#꼬꼬마 형태소 분석기(kKma())를 이용해 댓글에서 명사 추출
#꼬꼬마 형태소 분석기는 띄어쓰기 오류가 있는 문장에서도 형태소 잘 추출함.
#따라서, 정제되지 않은 텍스트(형태소) 분석할 때 적합.
```

```python
kkma = konlpy.tag.Kkma()
```

```python
#명사 추출 - apply()활용
nouns = df['reply'].apply(kkma.nouns)
```

```python
nouns #nouns는 리스트 자료구조 형태
```

```python
nouns = nouns.explode() #한 행에 한 단어가 들어가도록 구성
```

```python
nouns
```

```python
#nouns를 데이터프레임 자료구조형태로 바꾸고,
#두글자 이상으로 된 단어만 남기고
#단어빈도표 만들기
```

```python
#리스트 자료구조인 nouns를 데이터프레임 자료구조 형태로 바꾸기
df_word = pd.DataFrame({'word' : nouns})
```

```python
#글자 수 변수 추가
df_word['count'] = df_word['word'].str.len()
```

```python
#두 글자 이상인 단어만 남기기
df_word = df_word.query('count>=2')
```

```python
df_word
```

```python
#단어 빈도표 만들기
df_word = df_word.groupby('word', as_index = False)\
                .agg(n = ('word', 'count')) \
                .sort_values('n', ascending = False)
#(순서대로) 단어별 분리
#빈도 구하기
#내림차순 정렬
```

```python
df_word
```

```python
#댓글에 어떤 단어가 많이 사용됐는지 알아보기 쉽도록
#자주 사용된 상위 20개 단어를 추출해서 막대 그래프 만들기
```

```python
#자주 사용된 상위 20개 단어 추출
top20 = df_word.head(20)
```
