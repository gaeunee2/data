```python
# Q2
```

```python
df_midwest = df_midwest.rename(columns = {'poptotal':'total'})
```

```python
df_midwest = df_midwest.rename(columns = {'popasian':'asian'})
```

```python
df_midwest
```

```python
# Q3
```

```python
df_midwest['ratio'] = df_midwest['asian']/df_midwest['total'] * 100
```

```python
df_midwest['ratio'].plot.hist()
```

```python
#Q4
```

```python
df_midwest['ratio'].mean()
```

```python
df_midwest['size'] = np.where(df_midwest['ratio']>=df_midwest['ratio'].mean(), 'large','small')
```

```python
# Q5
```

```python
count_test = df_midwest['size'].value_counts()
```
