```python
#mpg데이터를 불러와서 복사본 만들기
import pandas as pd
```

```python
mpg = pd.read_csv('mpg.csv')
```

```python
mpg_copy = mpg.copy()
```

```python
#복사본 데이터를 이용해 cty는 city로, hwy는 highway로 수정하기
mpg_copy = mpg_copy.rename(columns = {'cty' : 'city'})
```

```python
mpg_copy = mpg_copy.rename(columns = {'hwy' : 'highway'})
```
