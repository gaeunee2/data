```python
mpg = pd.read_csv('mpg.csv')
```

```python
# 제조회사 및 구동방식별 분리 // drv는 구동방식 의미
# cty 평균 구하기 // cty는 도시연비 의미
mpg.groupby( ['manufacturer','drv']) \
    .agg(mean_cty = ('cty','mean'))
```

```python
#audi의 drv별 빈도 // drv는 구동방식 의미
mpg.query('manufacturer == "audi"') \
    .groupby(['drv']) \
    .agg(n = ('drv', 'count'))
```

```python
# chevrolet의 drv별 빈도 // drv는 구동방식 의미
mpg.query('manufacturer == "chevrolet"') \
    .groupby(['drv']) \
    .agg(n = ('drv','count'))
```

```python
# 제조 회사별로 
#'suv' 자동차의 도시 및 고속도로 합산 연비 평균을 구해 
# 내림차순으로 정렬하고
# 1~5위까지 출력하기
```
