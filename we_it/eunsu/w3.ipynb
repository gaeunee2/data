```python
import pandas as pd
df = pd.read_csv('news_comment_BTS.csv', encoding = 'UTF-8')
df.info()
```

```python
#불필요한 문자 제거하기
df['reply'] = df['reply'].str.replace('[^가-힣]',' ', regex = True)
df['reply'].head()
```

```python
#명사 추출
import konlpy
kkma = konlpy.tag.Kkma()
```

```python
nouns =df['reply'].apply(kkma.nouns)
nouns
```

```python
#한 핼에 한 단어가 들어가도록 구성
nouns = nouns.explode()
nouns
```

```python
#데이터 프레임 만들기
df_word = pd.DataFrame({'word' : nouns})

#글자 수 추가 
df_word['count'] = df_word['word'].str.len()

#두 글자 이상 단어만 남기기
df_word = df_word.query('count >= 2')
df_word
```

```python
#빈도표 만들기
df_word = df_word.groupby('word', as_index = False)\
.agg(n=('word','count'))\
.sort_values('n', ascending = False)

df_word
```

```python
#단어 빈도 상위 20개 추출
top20 = df_word.head(20)
top20
```

```python
#데이터 프레임을 딕셔너리로 변환
dic_word = df_word.set_index('word').to_dict()['n']
```

```python
# 필요한 라이브러리 import
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import PIL
import numpy as np

# 이미지 파일 열기
icon = PIL.Image.open('cloud.png')

# 마스크 이미지 생성 (흰색 배경에 아이콘을 붙여서)
img = PIL.Image.new('RGB', icon.size, (255, 255, 255))
img.paste(icon, icon)  # 아이콘을 새로운 이미지에 붙임
img = np.array(img)  # 이미지를 numpy 배열로 변환

# WordCloud 객체 생성
wc = WordCloud(random_state=1234,
               font_path=font,  # 사용할 글꼴 경로
               width=400,
               height=400,  # height를 40에서 400으로 수정
               background_color='white',
               mask=img)  # 마스크 이미지로 설정

# 단어 빈도수로 워드 클라우드 생성
img_wordcloud = wc.generate_from_frequencies(dic_word)

# 워드 클라우드 출력하기
plt.figure(figsize=(10, 10))  # 그림 크기 설정
plt.axis('off')  # 축 제거
plt.imshow(img_wordcloud, interpolation='bilinear')  # 생성된 워드 클라우드 표시
plt.show()  # 그래프 보여주기
```
